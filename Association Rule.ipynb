```python
#import libraries
import pandas as pd
import numpy as np
```

```python
# Read data
df = pd.read_csv(r"C:\Users\Arati\Downloads\Titanic.csv")
df
```

```python
#split data  (variable inside column then convert variables into different columns)
data = pd.get_dummies(df)
data
```

```python
#import libraries to get apriori values also association rules
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# by using apriori function we get minimum support criteria
score = apriori(data, min_support=0.2, use_colnames=True)
score
```

```python
# to get model 
model = association_rules(score,metric='lift')
```

```python
model
```

```python
# pandas filtering 
model[model['lift']>1]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df1 = pd.read_csv(r"C:\Users\Arati\Downloads\Supermarket.csv")
```

```python
df1
```

```python
# spliiting
def split(text):
    return text.split(',')
```

```python
# consider values
a = 'jam,butter,maggie'
split(a)
```

```python
# construct all the data into list form to convert it into columns
data = list(df1['Products'].apply(split))
data
```

```python
# to get values in the form of columns
te = TransactionEncoder()

en_df = te.fit_transform(data)
```

```python
#output
te.columns_
```

```python
en_df = pd.DataFrame(en_df, columns=te.columns_)
en_df
```

```python
# To convert data in categorical form
en_df.replace(False,0,inplace=True)
en_df.replace(True,1, inplace=True)
```

```python
en_df
```

```python
# find support score
score = apriori(en_df, min_support=0.2, use_colnames=True)
score
```

```python
# output
model = association_rules(score,metric='lift')
model
```
